# Types of Profit Margin 
- Gross profit margin : Focuses on production efficiency
- (revenue_cost of goods sold)*100/revenue
- b) Operating Profit Margin: Operating Profit ÷ Revenue × 100
- c) Net Profit Margin (most important): = Net Profit ÷ Revenue × 100

# GROSS, OPERATING, NET PROFIT MARGIN

```python
revenue= 1000000
cost = 700000
gorss_profit_margin = (revenue-cost)*100/revenue
print(gorss_profit_margin)
rent = 10000
salary=15000
operatingExpense=rent+salary
cogs=cost

operating_profit = revenue-operatingExpense-cogs
operating_profit_margin = round((operating_profit/revenue)*100)
print(f'operating expense: {operatingExpense}\nOperating Profit: {operating_profit}\nOperating Profit Margin: {operating_profit_margin}')

# calculate Net Revenue = Net Profit/revenue*100

profit=revenue-cost
tax =0.3
tax=profit*tax
total_expense = cogs+ operatingExpense+tax
Net_profit = revenue-total_expense
Net_profit_margin = (Net_profit/revenue)*100
print("Net Profit Margin: ",Net_profit_margin)
```

# 3. Classify profit into High Profit, Moderate Profit, or Loss based on value.

```python
revenues = [100000, 120000, 90000]
costs = [70000, 80000, 85000]

for i in range(len(revenues)):
    
    profit = revenues[i]-costs[i]

    if profit >= 30000:
        print('High Profit')
    elif profit >10000 and profit <30000:
        print('Moderate Profit')
    elif profit >0 and profit <10000:
        print('Low Profit')
    elif profit == 0:
        print('No Profit')
    else:
        print('Loss')
     
    
```

# 18. Find the number of companies with profit greater than a threshold value.

```python
companies = {
    "TCS": {
        "revenue": [100000, 120000, 90000],
        "cost": [70000, 80000, 85000]
    },
    "Infosys": {
        "revenue": [110000, 95000, 105000],
        "cost": [60000, 70000, 80000]
    }
}
print(companies)

max_profit=30000
count=0

for company, data in companies.items():
    revenue= data["revenue"]
    cost = data["cost"]

    for i, (r, c) in enumerate(zip(revenue, cost)):
        profit = r-c
        if profit > max_profit:
            count+=1



print(f'No of companies count above threshold value {count}')
```
